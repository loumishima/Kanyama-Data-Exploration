# Pandas using on Kanyama's dataset

### Importing dataset

```python
import pandas as pd
import numpy as np
import scipy as sp
import cv2
from skimage import io
```

```python
dados = pd.read_excel("/Users/gather3/Documents/Kanyama - Data Exploration/Kanyama Data Exploration/data/KANYAMA.XLSX", sheet_name= 1, skiprows=1)
dados.head()
```

```python
condition_column = [col for col in dados if col.startswith('Condit')]
photo_column = [photo for photo in dados if photo == "TAKE  PHOTO OF INSIDE THE TOILET" or photo == "TAKE PHOTO OF OUTSIDE THE TOILET/CONTAINMENT "]
dados = pd.concat((dados[photo_column], dados[condition_column]), axis = 1)
dados = dados.iloc[:,0:7].copy()
dados
```

```python
dados = dados.dropna(how = 'any')
dados.dtypes
```

```python
def removeSecondLinks(urls):
    final = []
    for link in urls :
        result = link.split("https")[1]
        result = 'https' + result
        final.append(result)
    return final
        
    
```

```python
dados.loc[:,'TAKE \xa0PHOTO OF INSIDE THE TOILET'] = removeSecondLinks(dados.loc[:,'TAKE \xa0PHOTO OF INSIDE THE TOILET'])
```

```python
def url_to_img(urls):
    result = []
    for url in urls:
        #print(f'downloading {url}')
        image = io.imread(url)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        width = int(image.shape[1] * 0.5)
        height = int(image.shape[0] * 0.5)
        image = cv2.resize(image, (width, height), interpolation = cv2.INTER_AREA)
        result.append(image)
    return result
        
```

```python
teste = url_to_img(dados.loc[0:15,'TAKE \xa0PHOTO OF INSIDE THE TOILET'])
#dados['TAKE \xa0PHOTO OF INSIDE THE TOILET']
```
