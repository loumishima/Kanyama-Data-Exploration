# Pandas using on Kanyama's dataset

### Importing dataset

```python
import pandas as pd
import numpy as np
import scipy as sp
import cv2
from skimage import io
```

```python
dados = pd.read_excel("/Users/gather3/Documents/Kanyama - Data Exploration/Kanyama Data Exploration/data/KANYAMA.XLSX", sheet_name= 1, skiprows=1)
```

```python
condition_column = [col for col in dados if col.startswith('Condit')]
condition_column = condition_column[0:5]
photo_column = [photo for photo in dados if photo == "TAKE  PHOTO OF INSIDE THE TOILET"]
rate = [col for col in dados if col.startswith('4.117')]
dados = pd.concat((dados[photo_column], dados[condition_column], dados[rate]), axis = 1)

dados.columns
```

```python
dados = dados.dropna(how = 'any')
dados.reset_index(inplace = True,drop = True)
```

```python
def removeSecondLinks(urls):
    final = []
    for link in urls :
        result = link.split("https")[1]
        result = 'https' + result
        final.append(result)
    return final
        
    
```

```python
dados.loc[:,'TAKE \xa0PHOTO OF INSIDE THE TOILET'] = removeSecondLinks(dados.loc[:,'TAKE \xa0PHOTO OF INSIDE THE TOILET'])
```

```python
def url_to_img(urls):
    result = []
    count = 0
    for url in urls:
        try:
            image = io.imread(url)
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
            width = int(image.shape[1] * 0.3)
            height = int(image.shape[0] * 0.3)
            image = cv2.resize(image, (width, height), interpolation = cv2.INTER_AREA)
            result.append(image)
            cv2.imwrite(f'/Users/gather3/Documents/Kanyama - Data Exploration/Kanyama Data Exploration/CNN - Images/{count}.jpg',
                       image)
            count += 1
        except:
            print(f'Link {url} is unavailable ')
            result.append(np.nan)
            count += 1
            pass
            
    return result
        
```

```python
#definitive_photos = url_to_img(dados.loc[:11,'TAKE \xa0PHOTO OF INSIDE THE TOILET'])
definitive_photos = url_to_img(dados.loc[:60,'TAKE \xa0PHOTO OF INSIDE THE TOILET'])
#dados['TAKE \xa0PHOTO OF INSIDE THE TOILET']
```

```python
np.save('definitive_photos', definitive_photos)

dados.loc[:60,'TAKE \xa0PHOTO OF INSIDE THE TOILET'] = definitive_photos

photo_prototype = dados.loc[:60, :].copy()

photo_prototype.head(60)
```
